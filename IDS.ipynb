```python
# ========== IMPORTS AND SETUP ==========
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import json
import pickle
import joblib
import time
from datetime import datetime
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.feature_selection import RFE, SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, GroupShuffleSplit
from sklearn.metrics import (accuracy_score, classification_report, 
                           confusion_matrix, roc_auc_score, roc_curve, 
                           precision_recall_curve, auc, f1_score, recall_score,
                           precision_score)
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import BayesianRidge
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTEENN
import tensorflow as tf
from tensorflow.keras.models import Model, Sequential, save_model
from tensorflow.keras.layers import (LSTM, Bidirectional, Dense, Dropout, 
                                   Input, concatenate, BatchNormalization,
                                   Conv1D, MaxPooling1D, Flatten, GlobalMaxPooling1D,
                                   Reshape, Embedding, Masking, Multiply, 
                                   GlobalAveragePooling1D, Attention, Add)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.utils.class_weight import compute_class_weight
import warnings
import random

# Bayesian Optimization
try:
    from skopt import BayesSearchCV
    from skopt.space import Real, Categorical, Integer
    BAYES_OPT_AVAILABLE = True
except ImportError:
    BAYES_OPT_AVAILABLE = False
    print("Bayesian optimization not available, falling back to random search")

warnings.filterwarnings('ignore')

# Configure GPU for TensorFlow
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"‚úÖ GPU configured: {len(gpus)} GPU(s) available")
    except RuntimeError as e:
        print(f"‚ùå GPU configuration failed: {e}")
        gpus = []
else:
    print("‚ùå No GPU found, using CPU")
    gpus = []

# Set reproducibility
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)
random.seed(seed)

print("Libraries imported successfully!")

# ========== IMPROVED DATA PREPROCESSOR WITH FEATURE SANITIZATION ==========
class ImprovedDataPreprocessor:
    """Improved preprocessor with feature sanitization and flow-based splitting"""
    
    def __init__(self, random_state=42):
        self.random_state = random_state
        self.label_encoders = {}
        self.scaler = StandardScaler()
        
    def preprocess_data(self, df, keep_datetime_for_sequences=False):
        """Main preprocessing function with feature sanitization"""
        print("üîÑ Starting Improved Data Preprocessing with Feature Sanitization...")
        
        # Create a copy
        df_processed = df.copy()
        
        # Separate features and target
        X = df_processed.drop(['Target', 'Traffic'], axis=1, errors='ignore')
        y = df_processed['Target']
        
        # 1. Feature Sanitization - Hash identifiers instead of dropping
        X = self._sanitize_features(X)
        
        # 2. Process datetime columns
        X = self._process_datetime_columns(X)
        
        # 3. Process categorical columns
        X = self._process_categorical_columns(X)
        
        # 4. Convert all to numeric
        X = self._convert_to_numeric(X)
        
        # 5. Handle missing values
        X = self._handle_missing_values(X)
        
        # 6. Scale features
        X_scaled = self.scaler.fit_transform(X)
        X = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)
        
        print(f"‚úÖ Improved Preprocessing completed. Final shape: {X.shape}")
        return X, y
    
    def _sanitize_features(self, X):
        """Sanitize features by hashing identifiers"""
        print("   üîí Sanitizing features (hashing identifiers)...")
        
        # Hash identifiers to preserve some structure without direct identification
        identifier_cols = ['SrcAddr', 'DstAddr']
        for col in identifier_cols:
            if col in X.columns:
                X[col] = X[col].apply(lambda x: hash(str(x)) % 10000)
        
        return X
    
    def _process_datetime_columns(self, X):
        """Process datetime columns and extract features"""
        datetime_cols = ['StartTime', 'LastTime']
        
        for col in datetime_cols:
            if col in X.columns:
                print(f"   Processing datetime column: {col}")
                try:
                    # Convert to datetime
                    X[col] = pd.to_datetime(X[col], format='%d-%m-%Y %H:%M', errors='coerce')
                    
                    # Extract features
                    X[f'{col}_hour'] = X[col].dt.hour
                    X[f'{col}_minute'] = X[col].dt.minute
                    X[f'{col}_dayofweek'] = X[col].dt.dayofweek
                    X[f'{col}_month'] = X[col].dt.month
                    X[f'{col}_day'] = X[col].dt.day
                    X[f'{col}_is_weekend'] = (X[col].dt.dayofweek >= 5).astype(int)
                    
                except Exception as e:
                    print(f"   ‚ö†Ô∏è Error processing {col}: {e}")
        
        # Calculate duration between StartTime and LastTime
        if 'StartTime' in X.columns and 'LastTime' in X.columns:
            try:
                X['duration_seconds'] = (X['LastTime'] - X['StartTime']).dt.total_seconds()
                X['duration_seconds'] = X['duration_seconds'].fillna(0).clip(lower=0)
            except:
                X['duration_seconds'] = 0
        
        # Remove original datetime columns
        X = X.drop(datetime_cols, axis=1, errors='ignore')
        
        return X
    
    def _process_categorical_columns(self, X):
        """Process categorical columns"""
        categorical_cols = ['Proto']
        
        for col in categorical_cols:
            if col in X.columns:
                print(f"   Encoding categorical column: {col}")
                try:
                    X[col] = X[col].fillna('unknown').astype(str)
                    le = LabelEncoder()
                    X[col] = le.fit_transform(X[col])
                    self.label_encoders[col] = le
                except Exception as e:
                    print(f"   ‚ö†Ô∏è Error encoding {col}: {e}")
                    # Simple hash encoding as fallback
                    X[col] = X[col].apply(lambda x: hash(str(x)) % 10000)
        
        return X
    
    def _convert_to_numeric(self, X):
        """Convert all columns to numeric"""
        for col in X.columns:
            if X[col].dtype == 'object':
                try:
                    X[col] = pd.to_numeric(X[col], errors='coerce')
                except:
                    X[col] = LabelEncoder().fit_transform(X[col].astype(str))
        return X
    
    def _handle_missing_values(self, X):
        """Handle missing values after all columns are numeric"""
        # Replace infinite values
        X = X.replace([np.inf, -np.inf], np.nan)
        
        # Simple imputation
        for col in X.columns:
            if X[col].isna().any():
                median_val = X[col].median()
                if pd.isna(median_val):
                    median_val = 0
                X[col] = X[col].fillna(median_val)
        
        return X

# ========== ADVANCED SEQUENTIAL DATA PREPARATION ==========
class SequentialDataBuilder:
    """Builds meaningful sequential data from network traffic"""
    
    def __init__(self, window_size=20, max_sequence_length=50):
        self.window_size = window_size
        self.max_sequence_length = max_sequence_length
        self.token_encoder = LabelEncoder()
        self.vocab_size = 0
        
    def build_flow_sequences(self, X, y):
        """Build sequences by grouping flows (SrcAddr, DstAddr, Proto)"""
        print("üîÑ Building flow-based sequences...")
        
        # Sort by time to maintain temporal order
        if 'StartTime_hour' in X.columns:
            time_cols = [col for col in X.columns if 'StartTime' in col or 'LastTime' in col]
            if time_cols:
                X_sorted = X.sort_values(by=time_cols)
                y_sorted = y.loc[X_sorted.index]
            else:
                X_sorted = X
                y_sorted = y
        else:
            X_sorted = X
            y_sorted = y
        
        # Create flow identifiers
        flow_columns = ['SrcAddr', 'DstAddr', 'Proto']
        available_flow_cols = [col for col in flow_columns if col in X_sorted.columns]
        
        if available_flow_cols:
            flow_ids = X_sorted[available_flow_cols].astype(str).apply('_'.join, axis=1)
        else:
            # Fallback: use index-based grouping
            flow_ids = (X_sorted.index // self.window_size).astype(str)
        
        sequences = []
        sequence_labels = []
        
        # Group by flow and create sequences
        for flow_id in flow_ids.unique():
            flow_indices = flow_ids[flow_ids == flow_id].index
            flow_data = X_sorted.loc[flow_indices]
            flow_labels = y_sorted.loc[flow_indices]
            
            # Create sequences within each flow
            for i in range(0, len(flow_data) - self.window_size + 1, self.window_size):
                sequence = flow_data.iloc[i:i + self.window_size].values
                label = flow_labels.iloc[i + self.window_size - 1]  # Label of last packet
                
                sequences.append(sequence)
                sequence_labels.append(label)
        
        if sequences:
            X_seq = np.array(sequences)
            y_seq = np.array(sequence_labels)
        else:
            # Fallback to simple windowing
            print("   ‚ö†Ô∏è Flow-based sequencing failed, using simple windowing")
            X_seq, y_seq = self._simple_windowing(X_sorted.values, y_sorted.values)
        
        print(f"‚úÖ Flow sequences: {X_seq.shape}")
        return X_seq, y_seq
    
    def build_token_sequences(self, X, y):
        """Build token-based sequences (NLP-style approach)"""
        print("üîÑ Building token-based sequences...")
        
        # Create tokens from key features
        token_features = ['SrcAddr', 'DstAddr', 'Proto', 'Sport', 'Dport']
        available_token_features = [col for col in token_features if col in X.columns]
        
        if not available_token_features:
            print("   ‚ö†Ô∏è No token features available, using flow-based sequences")
            return self.build_flow_sequences(X, y)
        
        # Create tokens
        tokens = []
        for _, row in X.iterrows():
            token_parts = []
            for col in available_token_features:
                token_parts.append(f"{col}_{row[col]}")
            tokens.append('_'.join(token_parts))
        
        # Encode tokens
        encoded_tokens = self.token_encoder.fit_transform(tokens)
        self.vocab_size = len(self.token_encoder.classes_)
        
        # Create sequences
        token_sequences = []
        sequence_labels = []
        
        for i in range(len(encoded_tokens) - self.window_size + 1):
            token_sequences.append(encoded_tokens[i:i + self.window_size])
            sequence_labels.append(y.iloc[i + self.window_size - 1])
        
        # Pad sequences
        X_seq = pad_sequences(token_sequences, maxlen=self.max_sequence_length, padding='post')
        y_seq = np.array(sequence_labels)
        
        print(f"‚úÖ Token sequences: {X_seq.shape}, Vocab size: {self.vocab_size}")
        return X_seq, y_seq
    
    def _simple_windowing(self, X, y):
        """Simple sliding window approach (fallback)"""
        sequences = []
        targets = []
        
        for i in range(len(X) - self.window_size):
            sequences.append(X[i:(i + self.window_size)])
            targets.append(y[i + self.window_size])
        
        return np.array(sequences), np.array(targets)

# ========== FLOW-BASED DATA SPLITTING ==========
def flow_based_split(X, y, test_size=0.2, random_state=42):
    """Split data by flow groups to prevent data leakage"""
    print("üîÑ Performing flow-based splitting...")
    
    # Create flow groups based on source and destination
    flow_columns = ['SrcAddr', 'DstAddr', 'Proto']
    available_flow_cols = [col for col in flow_columns if col in X.columns]
    
    if available_flow_cols:
        groups = X[available_flow_cols].astype(str).apply('_'.join, axis=1)
        print(f"   Created {len(groups.unique())} unique flow groups")
    else:
        # Fallback: use simple random split
        print("   ‚ö†Ô∏è No flow columns available, using random split")
        return train_test_split(X, y, test_size=test_size, random_state=random_state, stratify=y)
    
    # Perform group split
    gss = GroupShuffleSplit(n_splits=1, test_size=test_size, random_state=random_state)
    
    for train_idx, test_idx in gss.split(X, y, groups):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    print(f"   Train set: {X_train.shape}, Test set: {X_test.shape}")
    print(f"   Train groups: {len(groups.iloc[train_idx].unique())}, Test groups: {len(groups.iloc[test_idx].unique())}")
    
    return X_train, X_test, y_train, y_test

# ========== ADVANCED MODEL ARCHITECTURES WITH REGULARIZATION ==========
def create_regularized_lstm_model(input_shape, class_weights=None):
    """Create LSTM model with L2 regularization and dropout"""
    model = Sequential([
        LSTM(128, return_sequences=True, input_shape=input_shape,
             dropout=0.3, recurrent_dropout=0.3, 
             kernel_regularizer=l2(1e-4), recurrent_regularizer=l2(1e-4)),
        Dropout(0.3),
        BatchNormalization(),
        LSTM(64, dropout=0.3, recurrent_dropout=0.3,
             kernel_regularizer=l2(1e-4), recurrent_regularizer=l2(1e-4)),
        Dropout(0.3),
        BatchNormalization(),
        Dense(32, activation='relu', kernel_regularizer=l2(1e-4)),
        Dropout(0.4),
        Dense(16, activation='relu', kernel_regularizer=l2(1e-4)),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    
    lr = 0.001
    if class_weights and class_weights[1] > 2.0:
        lr = 0.0005
    
    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy', 'precision', 'recall', tf.keras.metrics.AUC(name='auc')]
    )
    return model

def create_tcn_attention_model(input_shape, class_weights=None):
    """Create Temporal Convolutional Network with Attention (Behavioral Regularization)"""
    inputs = Input(shape=input_shape)
    
    # Temporal Convolutional Layers
    x = Conv1D(64, 3, activation='relu', padding='same', kernel_regularizer=l2(1e-4))(inputs)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    
    x = Conv1D(128, 3, activation='relu', padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    
    # Attention mechanism
    attention = Dense(128, activation='sigmoid', kernel_regularizer=l2(1e-4))(x)
    x = Multiply()([x, attention])
    
    # Global pooling
    x = GlobalAveragePooling1D()(x)
    
    # Dense layers
    x = Dense(64, activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = Dropout(0.4)(x)
    x = BatchNormalization()(x)
    
    x = Dense(32, activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = Dropout(0.3)(x)
    
    outputs = Dense(1, activation='sigmoid')(x)
    
    model = Model(inputs=inputs, outputs=outputs)
    
    lr = 0.001
    if class_weights and class_weights[1] > 2.0:
        lr = 0.0005
    
    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy', 'precision', 'recall', tf.keras.metrics.AUC(name='auc')]
    )
    return model

def create_hybrid_cnn_lstm_model(input_shape, class_weights=None):
    """Create Hybrid CNN-LSTM model with regularization"""
    model = Sequential([
        Conv1D(64, 3, activation='relu', input_shape=input_shape, 
               kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        MaxPooling1D(2),
        Dropout(0.3),
        
        Conv1D(128, 3, activation='relu', kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        MaxPooling1D(2),
        Dropout(0.3),
        
        LSTM(64, dropout=0.3, recurrent_dropout=0.3,
             kernel_regularizer=l2(1e-4), recurrent_regularizer=l2(1e-4)),
        Dropout(0.4),
        BatchNormalization(),
        
        Dense(32, activation='relu', kernel_regularizer=l2(1e-4)),
        Dropout(0.3),
        Dense(16, activation='relu', kernel_regularizer=l2(1e-4)),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    
    lr = 0.001
    if class_weights and class_weights[1] > 2.0:
        lr = 0.0005
    
    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy', 'precision', 'recall', tf.keras.metrics.AUC(name='auc')]
    )
    return model

# ========== ADVANCED TRAINING WITH CLASS WEIGHTS ==========
def calculate_class_weights(y):
    """Calculate class weights for imbalanced datasets"""
    class_weights = compute_class_weight(
        'balanced',
        classes=np.unique(y),
        y=y
    )
    return {i: weight for i, weight in enumerate(class_weights)}

def train_with_class_weights(model, X_train, y_train, X_val, y_val, model_name, epochs=50):
    """Train model with class weights and proper callbacks"""
    class_weights = calculate_class_weights(y_train)
    
    callbacks = [
        EarlyStopping(patience=15, restore_best_weights=True, monitor='val_auc', mode='max'),
        ReduceLROnPlateau(factor=0.5, patience=8, min_lr=1e-6, monitor='val_loss')
    ]
    
    print(f"   üéØ Class weights: {class_weights}")
    
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=64,
        callbacks=callbacks,
        class_weight=class_weights,
        verbose=1
    )
    
    return history, class_weights

# ========== IMPROVED EVALUATION ==========
def evaluate_model_improved(model, X_test, y_test, model_name, model_type='tabular', history=None):
    """Improved model evaluation with detailed analysis"""
    print(f"\nüìä Evaluating {model_name}...")
    
    start_time = time.time()
    
    if model_type == 'sequential':
        y_proba = model.predict(X_test, verbose=0).flatten()
        y_pred = (y_proba > 0.5).astype(int)
    elif model_type == 'token_based':
        y_proba = model.predict(X_test, verbose=0).flatten()
        y_pred = (y_proba > 0.5).astype(int)
    else:
        if hasattr(model, 'predict_proba'):
            y_pred = model.predict(X_test)
            y_proba = model.predict_proba(X_test)[:, 1]
        else:
            y_pred = model.predict(X_test)
            y_proba = model.decision_function(X_test)
            # Normalize to [0, 1]
            y_proba = (y_proba - y_proba.min()) / (y_proba.max() - y_proba.min() + 1e-8)
    
    # Calculate comprehensive metrics
    accuracy = accuracy_score(y_test, y_pred)
    auc_score = roc_auc_score(y_test, y_proba)
    f1 = f1_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    
    # Class-wise recall
    recall_scores = recall_score(y_test, y_pred, average=None)
    recall_attack = recall_scores[1] if len(recall_scores) > 1 else recall_scores[0]
    recall_normal = recall_scores[0] if len(recall_scores) > 1 else 0
    
    # Latency measurement
    latency_start = time.time()
    if model_type in ['sequential', 'token_based']:
        _ = model.predict(X_test[:100], verbose=0)
        latency = (time.time() - latency_start) / 100 * 1000
    else:
        _ = model.predict(X_test[:1000])
        latency = (time.time() - latency_start) / 1000 * 1000
    
    # Print detailed results
    print(f"   ‚úÖ Accuracy: {accuracy:.6f}")
    print(f"   ‚úÖ AUC: {auc_score:.6f}")
    print(f"   ‚úÖ F1: {f1:.6f}")
    print(f"   ‚úÖ Precision: {precision:.6f}")
    print(f"   ‚úÖ Recall (Overall): {recall:.6f}")
    print(f"   ‚úÖ Recall (Attack): {recall_attack:.6f}")
    print(f"   ‚úÖ Recall (Normal): {recall_normal:.6f}")
    print(f"   ‚úÖ Latency: {latency:.4f} ms")
    
    return {
        'Model': model_name,
        'Accuracy': accuracy,
        'AUC': auc_score,
        'F1_Score': f1,
        'Precision': precision,
        'Recall': recall,
        'Recall_Attack': recall_attack,
        'Recall_Normal': recall_normal,
        'Latency_ms': latency,
        'Model_Type': model_type
    }, (y_test, y_pred, y_proba)

# ========== ADVANCED VISUALIZATION FUNCTIONS ==========
def create_comprehensive_visualizations(all_results, models_history, X_train_balanced, y_train_balanced, 
                                      X_train_original, y_train_original, results_dir):
    """Create comprehensive visualizations including confusion matrix collage"""
    print("\nüìä Creating comprehensive visualizations...")
    
    # Set style and font sizes
    plt.rcParams['font.size'] = 14
    plt.rcParams['axes.titlesize'] = 16
    plt.rcParams['axes.labelsize'] = 14
    plt.rcParams['xtick.labelsize'] = 12
    plt.rcParams['ytick.labelsize'] = 12
    plt.rcParams['legend.fontsize'] = 12
    
    # 1. SMOTE Before-After Comparison
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # Before SMOTE
    original_counts = np.bincount(y_train_original)
    axes[0].bar(['Normal', 'Attack'], original_counts, color=['blue', 'red'], alpha=0.7)
    axes[0].set_title('Class Distribution Before SMOTE')
    axes[0].set_ylabel('Count')
    for i, v in enumerate(original_counts):
        axes[0].text(i, v + max(original_counts)*0.01, str(v), ha='center', fontweight='bold', fontsize=16)
    
    # After SMOTE
    balanced_counts = np.bincount(y_train_balanced)
    axes[1].bar(['Normal', 'Attack'], balanced_counts, color=['lightblue', 'lightcoral'], alpha=0.7)
    axes[1].set_title('Class Distribution After SMOTE')
    axes[1].set_ylabel('Count')
    for i, v in enumerate(balanced_counts):
        axes[1].text(i, v + max(balanced_counts)*0.01, str(v), ha='center', fontweight='bold', fontsize=16)
    
    plt.tight_layout()
    plt.savefig(os.path.join(results_dir, 'smote_comparison.png'), dpi=300, bbox_inches='tight')
    plt.show()
    
    # 2. Training History Plots for Sequential Models
    if models_history:
        n_sequential = len(models_history)
        if n_sequential > 0:
            fig, axes = plt.subplots(2, 2, figsize=(20, 12))
            axes = axes.flatten()
            
            for idx, (model_name, history) in enumerate(models_history.items()):
                if idx >= 4:  # Limit to 4 plots
                    break
                    
                axes[idx].plot(history.history['loss'], label='Training Loss', linewidth=2)
                axes[idx].plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
                axes[idx].set_title(f'{model_name} - Loss', fontsize=16, fontweight='bold')
                axes[idx].set_xlabel('Epoch')
                axes[idx].set_ylabel('Loss')
                axes[idx].legend()
                axes[idx].grid(True, alpha=0.3)
            
            plt.tight_layout()
            plt.savefig(os.path.join(results_dir, 'training_history.png'), dpi=300, bbox_inches='tight')
            plt.show()
    
    # 3. Performance Comparison Radar Chart
    metrics = ['Accuracy', 'AUC', 'F1_Score', 'Recall_Attack', 'Precision']
    models = all_results['Model'].tolist()
    
    # Normalize metrics for radar chart
    normalized_data = []
    for metric in metrics:
        values = all_results[metric].values
        normalized = (values - values.min()) / (values.max() - values.min())
        normalized_data.append(normalized)
    
    # Create radar chart
    angles = np.linspace(0, 2*np.pi, len(metrics), endpoint=False).tolist()
    angles += angles[:1]  # Complete the circle
    
    fig, ax = plt.subplots(figsize=(12, 10), subplot_kw=dict(projection='polar'))
    
    colors = plt.cm.viridis(np.linspace(0, 1, len(models)))
    for i, model in enumerate(models):
        values = [data[i] for data in normalized_data]
        values += values[:1]  # Complete the circle
        ax.plot(angles, values, 'o-', linewidth=2, label=model, color=colors[i])
        ax.fill(angles, values, alpha=0.1, color=colors[i])
    
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(metrics, fontsize=14)
    ax.set_ylim(0, 1)
    ax.set_title('Model Performance Radar Chart', size=16, fontweight='bold', ha='center')
    ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0), fontsize=12)
    
    plt.savefig(os.path.join(results_dir, 'performance_radar.png'), dpi=300, bbox_inches='tight')
    plt.show()

def create_confusion_matrix_collage(all_predictions, results_dir):
    """Create a collage of confusion matrices for all models"""
    print("üìã Creating confusion matrix collage...")
    
    n_models = len(all_predictions)
    n_cols = min(3, n_models)
    n_rows = (n_models + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(6*n_cols, 5*n_rows))
    if n_models == 1:
        axes = np.array([axes])
    axes = axes.flatten()
    
    for idx, (model_name, (y_true, y_pred, y_proba)) in enumerate(all_predictions.items()):
        if idx >= len(axes):
            break
            
        cm = confusion_matrix(y_true, y_pred)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[idx],
                   cbar_kws={'shrink': 0.8}, annot_kws={"size": 24})  # 24 font size as requested
        
        axes[idx].set_title(f'{model_name}\nConfusion Matrix', fontsize=16, fontweight='bold')
        axes[idx].set_xlabel('Predicted', fontsize=14)
        axes[idx].set_ylabel('Actual', fontsize=14)
        axes[idx].set_xticklabels(['Normal', 'Attack'], fontsize=12)
        axes[idx].set_yticklabels(['Normal', 'Attack'], fontsize=12)
    
    # Hide empty subplots
    for idx in range(n_models, len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.savefig(os.path.join(results_dir, 'confusion_matrix_collage.png'), dpi=300, bbox_inches='tight')
    plt.show()

def create_classification_report_table(all_predictions, results_dir):
    """Create a comprehensive table of classification reports for all models"""
    print("üìä Creating classification report comparison table...")
    
    reports_data = []
    
    for model_name, (y_true, y_pred, y_proba) in all_predictions.items():
        report = classification_report(y_true, y_pred, output_dict=True)
        
        # Extract key metrics
        accuracy = accuracy_score(y_true, y_pred)
        auc_score = roc_auc_score(y_true, y_proba)
        
        # Class 0 (Normal)
        normal_precision = report['0']['precision']
        normal_recall = report['0']['recall']
        normal_f1 = report['0']['f1-score']
        normal_support = report['0']['support']
        
        # Class 1 (Attack)
        attack_precision = report['1']['precision']
        attack_recall = report['1']['recall']
        attack_f1 = report['1']['f1-score']
        attack_support = report['1']['support']
        
        # Weighted averages
        weighted_precision = report['weighted avg']['precision']
        weighted_recall = report['weighted avg']['recall']
        weighted_f1 = report['weighted avg']['f1-score']
        
        reports_data.append({
            'Model': model_name,
            'Accuracy': accuracy,
            'AUC': auc_score,
            'Normal_Precision': normal_precision,
            'Normal_Recall': normal_recall,
            'Normal_F1': normal_f1,
            'Normal_Support': normal_support,
            'Attack_Precision': attack_precision,
            'Attack_Recall': attack_recall,
            'Attack_F1': attack_f1,
            'Attack_Support': attack_support,
            'Weighted_Precision': weighted_precision,
            'Weighted_Recall': weighted_recall,
            'Weighted_F1': weighted_f1
        })
    
    # Create DataFrame and save
    reports_df = pd.DataFrame(reports_data)
    
    # Save as CSV
    reports_df.to_csv(os.path.join(results_dir, 'classification_reports_detailed.csv'), index=False)
    
    # Create a formatted table for display
    display_columns = ['Model', 'Accuracy', 'AUC', 'Attack_Precision', 'Attack_Recall', 'Attack_F1', 'Normal_Precision', 'Normal_Recall', 'Normal_F1']
    display_df = reports_df[display_columns].copy()
    
    # Format numbers
    for col in display_columns[1:]:
        display_df[col] = display_df[col].apply(lambda x: f'{x:.4f}')
    
    # Save formatted table
    display_df.to_csv(os.path.join(results_dir, 'classification_reports_formatted.csv'), index=False)
    
    # Create visual table
    fig, ax = plt.subplots(figsize=(16, max(6, len(display_df) * 0.6)))
    ax.axis('tight')
    ax.axis('off')
    
    table = ax.table(cellText=display_df.values,
                    colLabels=display_df.columns,
                    cellLoc='center',
                    loc='center',
                    bbox=[0, 0, 1, 1])
    
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1.2, 1.5)
    
    # Style header
    for i in range(len(display_df.columns)):
        table[(0, i)].set_facecolor('#40466e')
        table[(0, i)].set_text_props(weight='bold', color='white')
    
    plt.title('Comprehensive Classification Reports for All Models', 
              fontsize=16, fontweight='bold', pad=20)
    plt.savefig(os.path.join(results_dir, 'classification_reports_table.png'), 
                dpi=300, bbox_inches='tight')
    plt.show()
    
    return reports_df

# ========== ENHANCED PIPELINE ==========
def enhanced_pipeline():
    """Enhanced pipeline with flow-based splitting, regularization, and comprehensive visualization"""
    print("üöÄ ENHANCED IIoT NETWORK TRAFFIC CLASSIFICATION PIPELINE")
    print("="*80)
    
    # Create results directory
    results_dir = "/kaggle/working/enhanced-results"
    os.makedirs(results_dir, exist_ok=True)
    
    # Store all predictions for final visualization
    all_predictions = {}
    models_history = {}
    
    try:
        # 1. Load data
        print("üìä Loading dataset...")
        file_path = "/kaggle/input/wustl-iiot-2021-dataset/wustl_iiot_2021.csv"
        df = pd.read_csv(file_path)
        
        # Use smaller subset for testing
        df = df.head(1048576)  # Increased for better sequential learning
        
        print(f"Dataset shape: {df.shape}")
        target_counts = df['Target'].value_counts()
        print(f"Target distribution: {target_counts.to_dict()}")
        print(f"Attack ratio: {target_counts[1] / len(df):.4f}")
        
        # 2. Preprocess data with feature sanitization
        print("\nüîÑ Preprocessing data with feature sanitization...")
        preprocessor = ImprovedDataPreprocessor()
        X, y = preprocessor.preprocess_data(df)
        
        print(f"After preprocessing: {X.shape}")
        print(f"Features: {list(X.columns)}")
        
        # 3. Flow-based data splitting
        print("\nüîÑ Performing flow-based data splitting...")
        X_train, X_test, y_train, y_test = flow_based_split(X, y, test_size=0.2, random_state=42)
        
        # Store original distribution for SMOTE visualization
        X_train_original, y_train_original = X_train.copy(), y_train.copy()
        
        # 4. Balance training data with SMOTE
        print("\nüîÑ Applying SMOTE for class balancing...")
        smote = SMOTE(random_state=42)
        X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)
        
        print(f"Tabular data - Original train: {X_train.shape}, Balanced train: {X_train_balanced.shape}, Test: {X_test.shape}")
        
        # 5. Prepare sequential data
        print("\nüîÑ Preparing sequential data...")
        seq_builder = SequentialDataBuilder(window_size=20, max_sequence_length=30)
        
        # Flow-based sequences
        X_flow_seq, y_flow_seq = seq_builder.build_flow_sequences(X, y)
        
        if len(X_flow_seq) > 0:
            # Split sequential data using flow-based split
            X_flow_train, X_flow_test, y_flow_train, y_flow_test = flow_based_split(
                pd.DataFrame(X_flow_seq.reshape(X_flow_seq.shape[0], -1)), 
                pd.Series(y_flow_seq)
            )
            
            # Reshape back to sequential format
            X_flow_train = X_flow_train.values.reshape(-1, X_flow_seq.shape[1], X_flow_seq.shape[2])
            X_flow_test = X_flow_test.values.reshape(-1, X_flow_seq.shape[1], X_flow_seq.shape[2])
            y_flow_train = y_flow_train.values
            y_flow_test = y_flow_test.values
            
            # Balance sequential training data
            X_flow_train_flat = X_flow_train.reshape(X_flow_train.shape[0], -1)
            X_flow_train_balanced_flat, y_flow_train_balanced = smote.fit_resample(X_flow_train_flat, y_flow_train)
            X_flow_train_balanced = X_flow_train_balanced_flat.reshape(-1, X_flow_train.shape[1], X_flow_train.shape[2])
            
            print(f"Flow sequences - Train: {X_flow_train_balanced.shape}, Test: {X_flow_test.shape}")
        else:
            X_flow_train_balanced, y_flow_train_balanced, X_flow_test, y_flow_test = None, None, None, None
            print("   ‚ö†Ô∏è Flow sequences not available")
        
        # 6. Train and evaluate models
        all_results = []
        
        # Tabular models
        print("\nüîç TRAINING TABULAR MODELS...")
        tabular_models = {
            'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
            'LogisticRegression': LogisticRegression(random_state=42, max_iter=1000, n_jobs=-1),
            'SGD_Classifier': SGDClassifier(random_state=42, max_iter=1000, n_jobs=-1),
        }
        
        for name, model in tabular_models.items():
            print(f"   Training {name}...")
            model.fit(X_train_balanced, y_train_balanced)
            
            # Store predictions for final visualization
            results, predictions = evaluate_model_improved(model, X_test, y_test, name, 'tabular')
            all_results.append(results)
            all_predictions[name] = predictions
        
        # Sequential models (Flow-based)
        if X_flow_train_balanced is not None:
            print("\nüß† TRAINING SEQUENTIAL MODELS WITH REGULARIZATION...")
            sequential_models = [
                ('LSTM_Regularized', create_regularized_lstm_model((X_flow_train_balanced.shape[1], X_flow_train_balanced.shape[2]))),
                ('TCN_Attention', create_tcn_attention_model((X_flow_train_balanced.shape[1], X_flow_train_balanced.shape[2]))),
                ('Hybrid_CNN_LSTM', create_hybrid_cnn_lstm_model((X_flow_train_balanced.shape[1], X_flow_train_balanced.shape[2])))
            ]
            
            for name, model in sequential_models:
                print(f"   Training {name}...")
                try:
                    history, class_weights = train_with_class_weights(
                        model, X_flow_train_balanced, y_flow_train_balanced,
                        X_flow_test, y_flow_test, name, epochs=50
                    )
                    
                    # Store history for visualization
                    models_history[name] = history
                    
                    # Store predictions
                    results, predictions = evaluate_model_improved(model, X_flow_test, y_flow_test, name, 'sequential', history)
                    all_results.append(results)
                    all_predictions[name] = predictions
                    
                    # Save model
                    model.save(os.path.join(results_dir, f'{name.lower()}_model.h5'))
                    
                except Exception as e:
                    print(f"   ‚ùå {name} failed: {e}")
        
        # 7. Create comprehensive results table
        results_df = pd.DataFrame(all_results)
        results_df = results_df.sort_values('AUC', ascending=False)
        
        print("\nüìã COMPREHENSIVE RESULTS TABLE:")
        print("="*140)
        print(f"{'Model':<25} {'Type':<12} {'Accuracy':<10} {'AUC':<10} {'F1_Score':<10} {'Recall_Attack':<12} {'Latency_ms':<12}")
        print("="*140)
        
        for _, row in results_df.iterrows():
            model_type = row.get('Model_Type', 'tabular')
            print(f"{row['Model']:<25} {model_type:<12} {row['Accuracy']:.6f} {row['AUC']:.6f} {row['F1_Score']:.6f} "
                  f"{row['Recall_Attack']:.6f}    {row['Latency_ms']:.6f}")
        
        print("="*140)
        
        # 8. Create comprehensive visualizations
        create_comprehensive_visualizations(
            results_df, models_history, X_train_balanced, y_train_balanced,
            X_train_original, y_train_original, results_dir
        )
        
        # 9. Create confusion matrix collage and classification reports
        create_confusion_matrix_collage(all_predictions, results_dir)
        classification_reports_df = create_classification_report_table(all_predictions, results_dir)
        
        # 10. Save detailed results
        results_df.to_csv(os.path.join(results_dir, 'enhanced_results.csv'), index=False)
        results_df.to_json(os.path.join(results_dir, 'enhanced_results.json'), indent=4)
        
        # Save pipeline info
        pipeline_info = {
            'completion_time': datetime.now().isoformat(),
            'dataset_size': len(df),
            'best_model': results_df.iloc[0]['Model'],
            'best_auc': float(results_df.iloc[0]['AUC']),
            'best_accuracy': float(results_df.iloc[0]['Accuracy']),
            'models_tested': list(results_df['Model']),
            'improvements_applied': [
                'flow_based_splitting',
                'feature_sanitization_hashing',
                'l2_regularization',
                'dropout_regularization',
                'temporal_convolutional_networks',
                'attention_mechanisms',
                'early_stopping',
                'class_weight_balancing',
                'comprehensive_visualization_300dpi'
            ]
        }
        
        with open(os.path.join(results_dir, 'pipeline_info.json'), 'w') as f:
            json.dump(pipeline_info, f, indent=4)
        
        print(f"\nüéØ ENHANCED PIPELINE COMPLETED SUCCESSFULLY!")
        print(f"üèÜ BEST MODEL: {results_df.iloc[0]['Model']} "
              f"(AUC: {results_df.iloc[0]['AUC']:.6f}, "
              f"Accuracy: {results_df.iloc[0]['Accuracy']:.6f})")
        print(f"üìÅ Results saved to: {results_dir}")
        
        # Print key improvements
        print(f"\nüîß KEY IMPROVEMENTS APPLIED:")
        print(f"   ‚úÖ Flow-based splitting (no data leakage)")
        print(f"   ‚úÖ Feature sanitization (hashed identifiers)")
        print(f"   ‚úÖ L2 regularization and dropout")
        print(f"   ‚úÖ Temporal Convolutional Networks with Attention")
        print(f"   ‚úÖ Early stopping and class weights")
        print(f"   ‚úÖ Confusion matrix collage (24pt font)")
        print(f"   ‚úÖ Comprehensive classification reports table")
        print(f"   ‚úÖ SMOTE visualization graphs")
        print(f"   ‚úÖ Training/validation history graphs")
        print(f"   ‚úÖ All images saved at 300 DPI")
        
        return results_df, all_predictions, classification_reports_df
        
    except Exception as e:
        print(f"‚ùå Enhanced pipeline failed: {e}")
        import traceback
        traceback.print_exc()
        return None, None, None

# ========== EXECUTE ENHANCED PIPELINE ==========
if __name__ == "__main__":
    print("üöÄ Starting Enhanced Pipeline with All Improvements...")
    results, predictions, reports = enhanced_pipeline()
    
    if results is not None:
        print("\n" + "="*80)
        print("üéâ SUCCESS! ENHANCED PIPELINE COMPLETED")
        print("="*80)
        print("üìä Key Outputs Generated:")
        print("   ‚úÖ Confusion Matrix Collage (all models)")
        print("   ‚úÖ Classification Reports Comparison Table")
        print("   ‚úÖ SMOTE Before-After Visualization")
        print("   ‚úÖ Training/Validation History Graphs")
        print("   ‚úÖ Performance Radar Chart")
        print("   ‚úÖ All images saved at 300 DPI")
        print("   ‚úÖ Detailed CSV and JSON results")
        
        best_model = results.iloc[0]
        print(f"\nüèÜ BEST PERFORMING MODEL: {best_model['Model']}")
        print(f"üìä Best AUC: {best_model['AUC']:.6f}")
        print(f"üéØ Best Accuracy: {best_model['Accuracy']:.6f}")
        print(f"‚ö° Best F1 Score: {best_model['F1_Score']:.6f}")
        print(f"üîç Best Recall (Attack): {best_model['Recall_Attack']:.6f}")
        
        # Show file sizes
        results_dir = "/kaggle/working/enhanced-results"
        if os.path.exists(results_dir):
            print(f"\nüíæ Generated Files:")
            for file in os.listdir(results_dir):
                if file.endswith(('.png', '.csv', '.json')):
                    file_path = os.path.join(results_dir, file)
                    size_kb = os.path.getsize(file_path) / 1024
                    print(f"   üìÑ {file} ({size_kb:.1f} KB)")
    else:
        print("‚ùå Enhanced pipeline execution failed")
```
